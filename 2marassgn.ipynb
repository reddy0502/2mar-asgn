A scatter plot is a graphical representation of a set of data points. It is used to display the relationship between two variables, with each point in the plot representing a single data point. The horizontal axis represents one variable, and the vertical axis represents the other variable.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
```

The subplot() function is used in data visualization libraries like Matplotlib to create multiple plots in a single figure. This function allows you to divide a figure into a grid of rows and columns and specify the location of a plot .

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.subplot(2,2,4)
plt.plot(x4,y4)
```

A bar plot is a common type of chart that represents categorical data with rectangular bars. The height or length of each bar is proportional to the value it represents.

Bar plots are often used to compare the values of different categories or groups of data. They can be used to show changes in data over time, differences between groups, or the distribution of a single variable.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.barh(company,profit,color='b',edgecolor='black')
```

A box plot is a graphical representation of the distribution of a dataset.

Box plots are often used to show the range, skewness, and central tendency of a dataset, as well as any potential outliers or extreme values. They can also be used to compare the distribution of multiple datasets side-by-side.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
